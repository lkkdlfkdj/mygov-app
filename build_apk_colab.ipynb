```python
# 1. 安装系统依赖
!apt update -qq
!apt install -y -qq python3-pip python3-setuptools python3-dev \
  git zip unzip openjdk-17-jdk build-essential ccache \
  autoconf libtool pkg-config zlib1g-dev libncurses5-dev \
  libncursesw5-dev libtinfo5 cmake libffi-dev libssl-dev \
  libltdl-dev 2>&1 | tail -3
print('系统依赖安装完成')
```

```python
# 2. 安装 Python 工具
!pip install --upgrade pip setuptools wheel -q
!pip install buildozer cython virtualenv -q
print('Python 工具安装完成')
```

```python
# 3. 解压项目文件
import zipfile, os, glob

# 查找上传的 zip 文件
zips = glob.glob('mygov_app*.zip')
if not zips:
    # 尝试从默认上传路径找
    zips = glob.glob('/content/mygov_app*.zip')
if not zips:
    zips = glob.glob('*.zip')
    zips = [z for z in zips if 'mygov' in z.lower()]

if zips:
    zip_path = zips[0]
    print(f'找到: {zip_path}')
    with zipfile.ZipFile(zip_path, 'r') as zf:
        zf.extractall('mygov_app')
    os.chdir('mygov_app')
    print(f'项目文件: {len(os.listdir("."))} 项')
else:
    print('⚠ 未找到 mygov_app.zip')
    print('请先在左侧文件区（📁）上传 mygov_app.zip')
```

```python
# 4. 编译 APK（最耗时，约30-60分钟）
!python -m buildozer android debug 2>&1
print('\n编译完成！')
```

```python
# 5. 下载 APK
from google.colab import files
apk_list = glob.glob('bin/*.apk')
if apk_list:
    apk = apk_list[0]
    size_mb = os.path.getsize(apk) / 1024 / 1024
    print(f'APK 就绪: {apk} ({size_mb:.1f} MB)')
    files.download(apk)
else:
    print('未找到 APK 文件，编译可能失败')
    print('请检查上方输出日志')
```
